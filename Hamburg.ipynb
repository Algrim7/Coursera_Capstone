## Decision support for starting a new business in Hamburg

```python
import pandas as pd
pd.options.mode.chained_assignment = None
import numpy as np
import requests
from sklearn import preprocessing
from sklearn.cluster import KMeans
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import folium
```

### Clustering after age-distribution

#### Cleaning of the Data

```python
# Read data
df_age = pd.read_excel('StadtteilprofileBerichtsjahr2018.xlsx', skiprows=[0,1,2,3], header = None)

# Only important data in df
df_age = df_age[[0,3,5]]

# Rename columns
df_age = df_age.rename(columns={0: "Districts", 3: "Under_18", 5: "Over_65"})

# Remove Data that comes not from districts
df_age = df_age.drop(df_age.index[[0,12,16,27,33,39,40,47,83,91,99,100,101,102,103,104,105,106]])

# Change name of a few Districts so Foursquare founds its
df_age = df_age.replace(['Kleiner Grasbrook und Steinwerder', 'Waltershof und Finkenwerder', 'Altona-Nord', 'Neuland und Gut Moor', 'Moorburg und Altenwerder'], ['Steinwerder', 'Finkenwerder', 'Altona north', 'Neuland', 'Altenwerder'])

df_age.tail(10)
```

#### Group the data with k_means (k = 4)

```python
kclusters = 3

df_age_clustering = df_age.drop('Districts', 1)

kmeans = KMeans(n_clusters=kclusters, n_init = 100, random_state=0).fit(df_age_clustering)

df_age_cluster = df_age.copy()
df_age_cluster.insert(1, 'Cluster Labels', kmeans.labels_)
```

#### Analysis of the clusters

```python
df_age_cluster_grouped = df_age_cluster.groupby('Cluster Labels').mean().reset_index().set_index('Cluster Labels')
df_age_cluster_grouped
```

#### Clustering in three big groups

```python
np_age = df_age_cluster_grouped.to_numpy()
u18 = []
o65 = []

for i in np_age:
    u18 = u18 + [i[0]]
    o65 = o65 + [i[1]]

ind = np.arange(3) 
width = 0.35       
plt.bar(ind, u18, width, label='Under 18')
plt.bar(ind + width, o65, width, label='Over 65')

plt.ylabel('%')
plt.xlabel('Cluster')
plt.title('Comparison of the different Cluster')

plt.xticks(ind + width / 2, ('0', '1', '2'))
plt.legend(loc='best')
plt.savefig('age.png')
plt.show()
```

#### Function that loops through the Foursquare Data

```python
venues_list = []
VERSION = '20200101'
np_districts = df_age['Districts'].to_numpy()

for i in np_districts:
    NEAR = i
    url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&near={}&limit={}'.format(
        CLIENT_ID, 
        CLIENT_SECRET, 
        VERSION,
        NEAR,
        200)

    try:
        results = requests.get(url).json()["response"]['groups'][0]['items']
        venues_list.append([(
                    NEAR,
                    v['venue']['name'], 
                    v['venue']['categories'][0]['name']) for v in results])
    except: print(NEAR)

venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
venues.columns = ['District',
              'Venue', 
              'Venue_Category']
```

#### Delete Information that is not important for the analysis

```python
venues = venues[venues.Venue_Category != 'Supermarket']
venues = venues[venues.Venue_Category != 'Hotel']
venues = venues[venues.Venue_Category != 'Bus Stop']
venues = venues[venues.Venue_Category != 'Afghan Restaurant']
venues = venues[venues.Venue_Category != 'German Restaurant']
venues = venues[venues.Venue_Category != 'Greek Restaurant']
venues = venues[venues.Venue_Category != 'Eastern European Restaurant']
venues = venues[venues.Venue_Category != 'Italian Restaurant']
venues = venues[venues.Venue_Category != 'Turkish Restaurant']
venues = venues[venues.Venue_Category != 'Restaurant']
venues = venues[venues.Venue_Category != 'Café']
venues = venues[venues.Venue_Category != 'Seafood Restaurant']
venues = venues[venues.Venue_Category != 'Asian Restaurant']
venues = venues[venues.Venue_Category != 'Mexican Restaurant']
venues.shape
```

#### Onehotencoding

```python
hamburg_onehot = pd.get_dummies(venues[['Venue_Category']], prefix="", prefix_sep="")

hamburg_onehot['District'] = venues['District'] 

fixed_columns = [hamburg_onehot.columns[-1]] + list(hamburg_onehot.columns[:-1])
hamburg_onehot = hamburg_onehot[fixed_columns]

hamburg_grouped = hamburg_onehot.groupby('District').mean().reset_index()
hamburg_grouped.head()
```

#### Function that sortes the data

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

#### New Dataframe with the top 10 venues

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['District']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['District'] = hamburg_grouped['District']

for ind in np.arange(hamburg_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(hamburg_grouped.iloc[ind, :], num_top_venues)
    
neighborhoods_venues_sorted.head()
```

##### Group the data with k_means

```python
kclusters = 5

hamburg_grouped_clustering = hamburg_grouped.drop('District', 1)

kmeans = KMeans(n_clusters=kclusters, n_init = 100, random_state=0).fit(hamburg_grouped_clustering)

neighborhoods_venues_sorted_cluster = neighborhoods_venues_sorted.copy()
neighborhoods_venues_sorted_cluster.insert(1, 'Cluster Labels', kmeans.labels_)
```

#### Closer look to the interesting districts

```python
hamburg_onehot2 = pd.get_dummies(venues[['Venue_Category']], prefix="", prefix_sep="")

hamburg_onehot2['District'] = venues['District'] 

fixed_columns = [hamburg_onehot2.columns[-1]] + list(hamburg_onehot2.columns[:-1])
hamburg_onehot2 = hamburg_onehot2[fixed_columns]
hamburg_onehot2 = hamburg_onehot2.groupby('District').sum().reset_index()
hamburg_onehot2.head()
```

```python
df_Billwerder = hamburg_onehot2[hamburg_onehot2['District'] == 'Billwerder']
df_Billwerder = df_Billwerder.loc[:, (df_Billwerder != 0).any(axis=0)]
df_Billwerder = df_Billwerder.loc[:, (df_Billwerder != 1).any(axis=0)]
df_Billwerder = df_Billwerder.transpose().rename(columns={12: 'Billwerder'})
df_Billwerder.index.names = ['Venue']
df_Billwerder.reset_index(level=0, inplace=True)
df_Billwerder = df_Billwerder.drop(df_Billwerder.index[[0]])
df_Billwerder = df_Billwerder.sort_values(by=['Billwerder'])
ax = df_Billwerder.plot.barh(x='Venue', y='Billwerder', color = 'Blue',figsize=(10,8),fontsize=15,legend=False).get_figure().savefig('Billwerder.png')
```

```python
df_Curslack = hamburg_onehot2[hamburg_onehot2['District'] == 'Curslack']
df_Curslack = df_Curslack.loc[:, (df_Curslack != 0).any(axis=0)]
df_Curslack = df_Curslack.loc[:, (df_Curslack != 1).any(axis=0)]
df_Curslack = df_Curslack.transpose().rename(columns={17: 'Curslack'})
df_Curslack.index.names = ['Venue']
df_Curslack.reset_index(level=0, inplace=True)
df_Curslack = df_Curslack.drop(df_Curslack.index[[0]])
df_Curslack = df_Curslack.sort_values(by=['Curslack'])
ax = df_Curslack.plot.barh(x='Venue', y='Curslack', color = 'Green',figsize=(10,8),fontsize=15,legend=False).get_figure().savefig('Curslack.png')
```

```python
df_Sinstorf = hamburg_onehot2[hamburg_onehot2['District'] == 'Sinstorf']
df_Sinstorf = df_Sinstorf.loc[:, (df_Sinstorf != 0).any(axis=0)]
df_Sinstorf = df_Sinstorf.loc[:, (df_Sinstorf != 1).any(axis=0)]
df_Sinstorf = df_Sinstorf.transpose().rename(columns={70: 'Sinstorf'})
df_Sinstorf.index.names = ['Venue']
df_Sinstorf.reset_index(level=0, inplace=True)
df_Sinstorf = df_Sinstorf.drop(df_Sinstorf.index[[0]])
df_Sinstorf = df_Sinstorf.sort_values(by=['Sinstorf'])
ax = df_Sinstorf.plot.barh(x='Venue', y='Sinstorf', color = 'Red',figsize=(10,8),fontsize=15,legend=False).get_figure().savefig('Sinstorf.png')
```
